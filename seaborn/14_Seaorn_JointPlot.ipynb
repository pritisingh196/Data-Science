```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Drawing a first basic JOINT Plot*
sns.jointplot(data=iris,x='sepal_length',y='petal_length')
```

```python
# Change its kind to scatter|kde|hist|hex|reg|resid**
sns.jointplot(data=iris,x='sepal_length',y='petal_length'
             ,kind='resid')
```

```python
#. Grouping based on a categorical variable**
sns.jointplot(data=iris,x='sepal_length',y='petal_length'
             ,hue='species')
```

```python
# Formatting All and Individual Plots
sns.jointplot(data=iris,x='sepal_length',y='petal_length'
#              ,color='red'
#              ,hue='species'
#              ,palette='BuPu'
             ,joint_kws=dict(marker='+'
                            ,color='red')
             ,marginal_kws=dict(color='pink',kde=True))
```

```python
#  Adjust height, ratio, space and show/hide the marginal_ticks
sns.jointplot(data=iris,x='sepal_length',y='petal_length'
             ,height=5
             ,ratio=3
#              ,space=5
             ,marginal_ticks=True)
```

```python
# Plot KDE and RUG on top of Joint Plot
l=sns.jointplot(data=iris,x='sepal_length',y='petal_length',joint_kws=dict(color='red'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot,height=0.1,color='red')
```
