```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("/content/supermarket_sales.xlsx")
mart.head()
```

```python
## Create a basic BARPLOT**

plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',data=mart)
```

```python
#Add HUE to the barplot
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart)
```

```python
# Make the barplot HORIZONTAL
plt.figure(figsize=(15,5))

sns.barplot(x='Total',y='Product line',data=mart)
```

```python
# Plot the bars in a given ORDER**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female'])
```

```python
x=mart['Product line'].sort_values()
x.unique()
```

```python
# Add CAP on the ERROR BAR**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,capsize=0.2)
```

```python
# REMOVE the error bar using ci**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,ci=None)
```

```python
# Change bar colors using COLOR attribute**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,hue_order=['Male','Female']
           ,color='red')
```

```python
# ** Change color using PALLETE attribute**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis')
```

```python
# **Using SATURATION parameter**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis'
           ,saturation=0.1)
```

```python
# **CHANGE DEFAULT AGGREGATION METHOD USING ESTIMATOR PARAMETER**
plt.figure(figsize=(15,5))

sns.barplot(x='Product line',y='Total',hue='Gender',data=mart,order=['Electronic_accessories', 'Fashion_accessories','Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle','Sports_and_travel']
           ,palette='cividis'
           ,estimator=np.median)
```
