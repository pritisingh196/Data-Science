```python
#markersize

import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3,4,7,9])

plt.plot(ypoints, marker = 'o', ms = 25)
plt.show()
```

```python
# markeredgecolor

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 16, 10])

plt.plot(ypoints, marker = 'o', ms = 25,mec='r')
plt.show()
```

```python
# markerfacecolor

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 7, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 25,mfc='g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 16, 10])

plt.plot(ypoints, marker = 'o', ms = 25,mec='hotpink',mfc='hotpink')
plt.show()
```
