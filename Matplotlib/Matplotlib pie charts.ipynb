```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
#label
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
label=["may","june","july","aug"]
plt.pie(y,labels=label)
plt.show() 
```

```python
#angle
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
label=["may","june","july","aug"]
plt.pie(y,labels=label,startangle=90)
plt.show() 
```

```python
#explode
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
label=["may","june","july","aug"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=label,explode=myexplode)
plt.show() 
```

```python
#shadow
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["April", "may", "june", "july"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
#colors

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["April", "may", "june", "july"]
mycolors = ["yellow","orange","blue","red"]

plt.pie(y, labels = mylabels,colors=mycolors)
plt.show()
```

```python
#legends
#colors

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["April", "may", "june", "july"]
mycolors = ["yellow","orange","blue","red"]

plt.pie(y, labels = mylabels,colors=mycolors)
plt.legend()
plt.show()
```

```python
#legends
#colors
#title

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["April", "may", "june", "july"]
mycolors = ["yellow","orange","blue","red"]

plt.pie(y, labels = mylabels,colors=mycolors)
plt.legend(title="months")
plt.show()
```
