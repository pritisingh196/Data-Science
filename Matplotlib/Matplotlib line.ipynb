```python
# line styles
import matplotlib.pyplot as plt
import numpy as np
x=([10,16,30,35,50])
plt.plot(x,linestyle='dotted')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=([10,16,30,35,50])
plt.plot(x,linestyle='dashed')
plt.show()
```

```python
#line color
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([5, 8, 1, 10])

plt.plot(ypoints, color = 'y')
plt.show()
```

```python
#line color
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([5, 8, 1, 10])

plt.plot(ypoints, c = 'hotpink')
plt.show()
```

```python
#line color
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([5, 8, 1, 10])

plt.plot(ypoints, linewidth = '15.5',c='g')
plt.show()
```

```python
#multiple lines 
import matplotlib.pyplot as plt
import numpy as np

x= np.array([3, 8, 1, 10])
y= np.array([6, 2, 7, 11])

plt.plot(x)
plt.plot(y)

plt.show()
```
