```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 3, 10])

plt.bar(x,y)
plt.show()
```

```python
#horizontal bars
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 3, 10])

plt.barh(x,y)
plt.show()
```

```python
# color
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 3, 10])

plt.barh(x,y,color='yellow')
plt.show()
```

```python
#bar width
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 3, 10])

plt.bar(x,y, width=0.3)
plt.show()
```
